```python
import pandas as pd
import numpy as np
```

# Primera Parte

Los datos de https://www.swpc.noaa.gov/products/solar-cycle-progression sobre el flux 10.7 empiezan desde octubre del 2004, pero hay en https://lasp.colorado.edu/lisird/data/penticton_radio_flux hay datos desde 1947, por lo que primero voy a juntar estos datos para obtener el 10.7 desde 1947 hasta 2025:

## 1947

```python
val = 'flux'
```

```python
df_47 = pd.read_csv("Flux_1947_1996.txt")
df_47.head()
```

```python
pd.to_numeric(df_47["Adj."])
```

```python
df_47.drop(columns=['Julian Date', 'Rotation'], inplace=True)
df_47.head()
```

```python
df_47.replace(0, np.nan, inplace=True)

df_47.head()
```

```python
df_47.rename(columns={"mo": "month", "dy": "day", "Adj.": val}, inplace=True)
df_47['Date'] = pd.to_datetime(df_47[['year','month','day']])
df_47.drop(columns=['year', 'month', 'day'], inplace=True)
df_47 = df_47[['Date', val]]
df_47.head()
```

```python
dat_47 = pd.DataFrame(df_47.groupby(pd.PeriodIndex(df_47['Date'], freq="M"))[val].mean())
dat_47
```

## 1996

```python
df_96 = pd.read_csv("Flux_1996_2004.txt", delimiter="  ")
df_96.head()
```

```python
df_96.drop(columns=['Julian Day', 'Rotation', 'U.T.', 'xx', 'xx.1', 'xx.2'], inplace=True)
df_96.head()
```

```python
df_96.rename(columns={"Mo": "month", "Dy": "day", "Adjusted": val}, inplace=True)
df_96['Date'] = pd.to_datetime(df_96[['Year','month','day']])
df_96.drop(columns=['Year', 'month', 'day'], inplace=True)
df_96 = df_96[['Date', val]]
df_96.head()
```

```python
dat_96 = pd.DataFrame(df_96.groupby(pd.PeriodIndex(df_96['Date'], freq="M"))[val].mean())
dat_96
```

## 2004

```python
df_04 = pd.read_csv("Flux_2004.txt", skiprows=[1], sep=r'\s+')
df_04.head()
```

```python
df_04.drop(columns=['fluxjulian', 'fluxcarrington', 'fluxtime'], inplace=True)
df_04.head()
```

```python
df_04.rename(columns={"fluxadjflux": val}, inplace=True)
df_04['Date'] = pd.to_datetime(df_04['fluxdate'], format='%Y%m%d')
df_04.drop(columns=['fluxdate'], inplace=True)
df_04 = df_04[['Date', val]]
df_04.head()
```

```python
dat_04 = pd.DataFrame(df_04.groupby(pd.PeriodIndex(df_04['Date'], freq="M"))[val].mean())
dat_04
```

## Uniendo los DataFrames

```python
dat_47.drop(index=['1996-02', '1996-03'], inplace=True)
```

```python
dat_96.drop(dat_96.loc['2004-10':].index, inplace=True)
```

```python
Datos = [dat_47, dat_96, dat_04]
```

```python
data = pd.DataFrame(pd.concat(Datos))
```

```python
mod_data = data.round(decimals=1)*10
mod_data.loc["2025"]
```

# Segunda Parte: Uniendo Manchas Solares y Flux 10.7

```python
df_swpc = pd.read_json("Solar-Cycle-Data.json")
df_swpc
```

```python
df_swpc.drop(columns=["smoothed_f10.7", "f10.7", "smoothed_swpc_ssn", "smoothed_ssn"], inplace=True)
df_swpc.replace(-1, np.nan, inplace=True)
df_swpc
```

```python
df_swpc['Date'] = pd.to_datetime(df_swpc['time-tag'], format='%Y-%m').dt.strftime('%Y-%m')
df_swpc.drop(columns=["time-tag"], inplace=True)
df_swpc["Date"] = pd.to_datetime(df_swpc["Date"])
df_swpc["Date"] = df_swpc["Date"].dt.to_period("M")
df_swpc.set_index(df_swpc['Date'], inplace=True)
df_swpc.drop(columns=["Date"], inplace=True)
df_swpc
```

```python
total_data = df_swpc.join(data)
total_data.rename(columns={"observed_swpc_ssn": "swpc_ssn"}, inplace=True)
total_data
```

```python
total_data.to_csv("Sun_Data.csv")
```
